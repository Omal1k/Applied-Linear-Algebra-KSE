## Transformations for 2D objects

```python
class Transformers2D():
    def __init__(self, origin_matrix, marker='Почтакова матриця '):
        self.origin_matrix = origin_matrix
        self.marker = ' ' + marker
    
    def rotate(self, angle): 
        radians = np.radians(angle)
        rotate = np.array([[np.cos(radians), -np.sin(radians)],
                          [np.sin(radians), np.cos(radians)]])
        return Transformers2D(self.origin_matrix @ rotate, marker=self.marker + f'повернтий на {angle} градусів; ')
    
    def scaling(self, k_x=1, k_y=1):
        transform = np.array([[k_x, 0],
                             [0, k_y]])
        return Transformers2D(self.origin_matrix @ transform, marker=f'збільшений у {k_x} по 0x і у {k_y} по 0y; ')
    
    def reflection_x(self):
        transform = np.array([[1, 0],
                             [0, -1]])
        
        return Transformers2D(self.origin_matrix @ transform, marker=f'Об\'єкт відзеркалений відносно вісі 0x; ')
    
    def reflection_y(self):
        transform = np.array([[-1, 0],
                             [0, 1]])
        
        return Transformers2D(self.origin_matrix @ transform, marker=f'Об\'єкт відзеркалений відносно вісі 0y; ')
    
    def move_x(self, angle):
        radians = np.radians(angle)
        transform = np.array([[1, np.tan(radians)],
                             [0, 1]])
        return Transformers2D(self.origin_matrix @ transform, marker=f'Нахил осі координат на кут 0x на {angle}; ')
    
    def move_y(self, angle):
        radians = np.radians(angle)
        transform = np.array([[1, 0], 
                              [np.tan(radians), 1]])
        return Transformers2D(self.origin_matrix @ transform, marker=f'ОНахил осі координат на кут 0y на {angle}; ')
    
    def any_transforamtion(self, matrix): # будь-яка матрциця перетворення
        return Transformers2D(self.origin_matrix @ matrix, marker= 'довільне перетворення; ')
    
    def visualizatin(self): # візуалізація
        plt.plot(self.origin_matrix[:, 0], self.origin_matrix[:, 1], 'r-', color='green')

        plt.axhline(0, color='black',linewidth=0.5)
        plt.axvline(0, color='black',linewidth=0.5)
        plt.grid(color = 'gray', linestyle = '-', linewidth = 0.5)
        plt.gca().set_aspect('equal', adjustable='box')
        plt.title(self.marker)
        plt.show()
        
        
```

```python
trans = Transformers2D(batman)
```

```python
sth.scaling(2, 2).rotate(45).visualizatin()
```

## Transformations for 3D figures

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
class Transformers3D():
    def __init__(self, origin_matrix):
        self.origin_matrix = origin_matrix
        
    def rotate(self, angle, axis): 
        radians = np.radians(angle)
        
        rotate_x = np.array([[1, 0, 0],
                           [0, np.cos(radians), -np.sin(radians)],
                           [0, np.sin(radians), np.cos(radians)]])
        
        rotate_y = np.array([[np.cos(radians), 0, np.sin(radians)],
                           [0, 1, 0],
                           [-np.sin(radians), 0, np.cos(radians)]])
        
        rotate_z = np.array([[np.cos(radians), -np.sin(radians), 0],
                           [np.sin(radians), np.cos(radians), 0],
                           [0, 0, 1]])
        
        if axis=='x':
            return Transformers3D(self.origin_matrix @ rotate_x)
        elif axis=='y':
            return Transformers3D(self.origin_matrix @ rotate_y)
        elif axis=='z':
            return Transformers3D(self.origin_matrix @ rotate_z)
        
    def scaling(self, k_x, k_y, k_z):
        transform = np.array([[k_x, 0, 0],
                             [0, k_y, 0],
                             [0, 0, k_z]])
        return Transformers3D(self.origin_matrix @ transform)
    
    
    def plot3d(self, label):
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        x = self.origin_matrix[:, 0]
        y = self.origin_matrix[:, 1]
        z = self.origin_matrix[:, 2]
        ax.plot(x, y, z, marker='o')
        ax.set_title(label)
        plt.show()
```

```python
d3 = Transformers3D(np.array([
    [0, 0, 0], [1, 0, 0], [0.5, 0.5, 1],
    [0, 0, 0], [0, 1, 0], [0.5, 0.5, 1],
    [0, 0, 0], [1, 1, 0], [0.5, 0.5, 1],
    [1, 0, 0], [1, 1, 0], [0.5, 0.5, 1],
    [0, 1, 0], [1, 1, 0], [0.5, 0.5, 1]
]))
scaled_3d = d3.scaling(2, 2, 2)
rotated_x = scaled_3d.rotate(30, 'x')
```

```python
import torch
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import cv2 as cv

image_path = 'img/cat.png'
image = cv.imread(image_path)

image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

transform_to_tensor = transforms.ToTensor()
tensor_image = transform_to_tensor(image)

def show_tensor_image(tensor_img):
    transform_to_pil = transforms.ToPILImage()
    pil_img = transform_to_pil(tensor_img)
    plt.imshow(pil_img)
    plt.axis('off')
    plt.show()

show_tensor_image(tensor_image)
```

```python
rotate_transform = transforms.RandomRotation(degrees=(90, 90))
scaled_transform = transforms.Resize((512, 1024))

rotated_image = rotate_transform(transforms.ToPILImage()(tensor_image))
tensor_rotated_image = transform_to_tensor(rotated_image)

scaled_image = scaled_transform(transforms.ToPILImage()(tensor_image))
tensor_scaled_image = transform_to_tensor(scaled_image)

print("Обертання на 90 градусів:")
show_tensor_image(tensor_rotated_image)

print("Масштабування до 256x256:")
show_tensor_image(tensor_scaled_image)
```
